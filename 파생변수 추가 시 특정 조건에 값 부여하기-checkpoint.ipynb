```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# dataframe 형태로 변환해서 분석 시작.
import numpy as np
import pandas as pd

# feature names와 target을 record로 갖는 데이터프레임 생성.
```

```python
data = pd.DataFrame(data=iris.data,columns=iris.feature_names)
data['target'] = iris.target
data
```

```python
# target 값 0 : setosa, 1 : versicolor, 2:'virginica'
data['target'] = data['target'].map({0:'setosa',1:'versicolor',2:'virginica'})
data
```

```python
data.describe()
```

```python
# 파생변수 추가.
data['total'] = data['sepal length (cm)']*data['sepal width (cm)']*data['petal length (cm)']*data['petal width (cm)']
```

```python
# total 값이 93 이하 : 1, 163 이하 : 2, 163 이상 : 3 부여
```

```python
data['total_category'] = 0

data.loc[data['total']<=93,'total_category']=1
data.loc[(data['total']>93) & (data['total']<=163) ,'total_category']=2
data.loc[data['total']>163,'total_category']=3
data.head()
```

```python
data['total_category'].value_counts()
```
