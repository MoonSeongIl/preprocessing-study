## Label Encoder 만들기

```python
# scikit-learn의 sample data load.
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# dataframe 형태로 변환해서 분석 시작.
import numpy as np
import pandas as pd

# feature names와 target을 record로 갖는 데이터프레임 생성.

data = pd.DataFrame(data=iris.data,columns=iris.feature_names)
data['target'] = iris.target
data
```

```python
# target 값 0 : setosa, 1 : versicolor, 2:'virginica'
data['target'] = data['target'].map({0:'setosa',1:'versicolor',2:'virginica'})
data
```

```python
# object 형태의 데이터를 Label Encoder를 통해서 변환.

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

c = data.columns[data.dtypes=='object']
c
```

```python
for i in c:
    data[i] = le.fit_transform(data[i])
    
data
```
